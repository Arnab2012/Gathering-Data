```python
# creating empty dataframe
df=pd.DataFrame()
```

```python
# importing the total dataset using loop
for i in range (1,541):
    response=requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=6649e6e9c477bc4c316b49e90b94a80f&language=en-US&page={}'.format(i))
    temp_df=pd.DataFrame(response.json()['results'])[['id','popularity','title','vote_count']]
    df=df.append(temp_df,ignore_index=True)
```
